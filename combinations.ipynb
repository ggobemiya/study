```python
def combination(arr, r):
    # 1.
    arr = sorted(arr)
    used = [0 for _ in range(len(arr))]

    # 2.
    x0 = 0
    def generate(chosen, x0):
        if len(chosen) == r:
            print('=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=')
            print('ANSWER : ', chosen)
            print('-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-')
            return

    	# 3.
        start = arr.index(chosen[-1]) + 1 if chosen else 0
        print('chosen, start :', chosen, start)
        
        for nxt in range(start, len(arr)):
            print('    ','for문 시작 : nxt =',nxt, 'in range:(',start,'~', len(arr),')')
            if used[nxt] == 0 and (nxt == 0 or arr[nxt-1] != arr[nxt] or used[nxt-1]):
                print('nxt=',nxt)
                chosen.append(arr[nxt])
                used[nxt] = 1
                x0 +=1
                print('        ',x0, '차 재귀함수 진입')
                generate(chosen, x0)
                print('        ',x0, '차 재귀함수 탈출')
                x0 -=1
                chosen.pop()
                used[nxt] = 0
    generate([], x0)
```
